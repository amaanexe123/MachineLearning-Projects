### Separating the legitimate and Fraudulent transactions

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
print(legit.shape)
print(fraud.shape)
```

### Dealing with the unbalanced data using under-sampling. Under sampling is building a sample dataset from the original dataset containing similar distribution of normal transactions and fraudulent transactions

### Number of fraudulent transactions = 492

```python
legit_sample = legit.sample(n=492)
```

### Concatenating two dataframes, legit_sample and fraud. So now we will have equal amount of transactions in the new dataset

```python
new_data = pd.concat([legit_sample, fraud],axis = 0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

### Splitting the dataset into independent and independent variables

```python
X = new_data.iloc[:, :-1].values
Y = new_data.iloc[:, -1].values
print(X)
print(Y)
```

### Splitting into Training Data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

### Using Logistic Regression

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

### Checking the accuracy of the model on training data

```python
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
print(f'The accuracy on training data is {training_data_accuracy * 100}%')
```

### Checking the accuracy of the model on test data

```python
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
print(f'The accuracy on test data is {test_data_accuracy * 100}%')
```

### This model has a good accuracy of 94% on both training and test data, and can be used to detect legit and fradulent credit card transactions
